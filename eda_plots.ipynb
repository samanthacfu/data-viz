# Common EDA Plots

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
## This is mostly for my reference since seaborn is so extensively documented
## https://seaborn.pydata.org/introduction.html

iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.set(style='white', 
        font_scale=0.7,
        font = 'Arial',
        palette = 'husl'
        )
mpl.rcParams['figure.dpi']= 130
```

## Scatter Plot

```python
ax = sns.scatterplot(x='sepal_length', 
                     y='sepal_width', 
                     hue='species', 
                     style='species',
                     #size='petal_length',
                     data=iris);

sns.despine()

ax.set(xlabel='Sepal Length', 
       ylabel='Sepal Width')

plt.ylim(0, None)
plt.xlim(0, None)
```

## Line of Best Fit 

```python
ax = sns.lmplot(x='sepal_length',
                y='sepal_width',
                hue='species',
                data=iris)

sns.despine()

ax.set(xlabel='Sepal Length', 
       ylabel='Sepal Width')

#plt.ylim(0, None)
#plt.xlim(0, None)
```

## Pair Plots

```python
sns.pairplot(data=iris, hue="species")
```

## Histogram

```python
unique_species = iris.species.unique().tolist()
```

```python
for species in unique_species: 
    ax = sns.distplot(iris.sepal_length[iris.species==species], label=species)
sns.despine()

ax.set_title('Distribution of Sepal Length Between Species', fontsize=14)
ax.set_xlabel("Frequency",fontsize=10)
ax.set_ylabel("Sepal Length",fontsize=10)
ax.legend(loc='center left', frameon=True)
plt.ylim(0, None)
plt.xlim(0, None)
```

## Violin Plots

```python
## Just because I think these look cool

ax = sns.violinplot(x="species", y="sepal_length", data=iris)
sns.despine()
ax.set_ylabel("Sepal Length",fontsize=10)
ax.set_xlabel("Species",fontsize=10)
```
